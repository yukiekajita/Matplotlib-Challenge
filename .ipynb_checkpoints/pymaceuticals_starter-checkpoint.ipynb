## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
# Combine the data into a single dataset
combined_data = pd.merge(study_results, mouse_metadata, on="Mouse ID", how="outer")
```

```python
# Reorganize columns
organized_combined_data = combined_data[["Mouse ID", "Timepoint", "Tumor Volume (mm3)", 
                                         "Metastatic Sites", "Drug Regimen", "Sex", "Age_months", "Weight (g)"]]
```

```python
# Sort data with Timepoint
organized_combined_data = combined_data.sort_values(["Timepoint"])
```

```python
# Check the number of mice and dataset
organized_combined_data.describe()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.  
duplicated_data = organized_combined_data[organized_combined_data.duplicated(["Mouse ID", "Timepoint"], keep=False)]

# Optional: Get all the data for the duplicate mouse ID. 
duplicated_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
new_organized_combined_data = organized_combined_data.drop_duplicates(["Mouse ID", "Timepoint"], keep='first')
```

```python
# Check the number of mice in the clean DataFrame and dataset after removing the duplicated data
new_organized_combined_data.describe()
```

```python
# Final Version of Clean Dataset with Reset Index numbers
new_index_data = new_organized_combined_data.reset_index(drop=True)
```

```python
# Display the data table for preview
new_index_data.head()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward: creating multiple series and 
means_tumor_volume = new_index_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
median_tumor_volume = new_index_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
var_tumor_volume = new_index_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
std_tumor_volume = new_index_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
SEM_tumor_volume = new_index_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()

# Put them all together and make Dataframe
summary_statistics_df = pd.DataFrame({"Mean Tumor Volume": means_tumor_volume, "Median Tumor Volume":median_tumor_volume, 
                                     "Tumor Volume Variance": var_tumor_volume, "Tumor Volume Std. Dev.":std_tumor_volume, 
                                     "Tumor Volume Std. Err.": SEM_tumor_volume})
summary_statistics_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Producing everything in a single groupby function with .agg
alt_summary_statistics_df = new_index_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(['mean', 'median', 'var', 'std', 'sem'])
alt_summary_statistics_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume (max tumor) of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Obtain the final tumor volume (max tumor) of each mouse with groupby ID and timepoint max
max_tumor = new_index_data.groupby(["Mouse ID"])['Timepoint'].max()

# Put a new index number for the max tumor dataset
max_tumor = max_tumor.reset_index()
```

```python
# Merge max_tumor with the original dataframe to get the tumor volume at the last timepoint
merged_data = pd.merge(max_tumor, new_index_data, on=['Mouse ID','Timepoint'], how="left")
merged_data 
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_list = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in treatment_list:

    # Locate the rows which contain mice on each drug and get the tumor volumes 
    final_tumor_vol = merged_data.loc[merged_data["Drug Regimen"] == drug, 'Tumor Volume (mm3)']
    
    # add subset to the tumor_vol_list   
    tumor_vol_list.append(final_tumor_vol)
    
    # Calculate quartile for each drug regimen
    quartiles = final_tumor_vol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq-(1.5*iqr)
    upper_bound = upperq+(1.5*iqr)
    
    # Obtain outliers and print out for each drug regimen
    outliers = final_tumor_vol.loc[(final_tumor_vol < lower_bound) | (final_tumor_vol > upper_bound)]
    print(f"{drug}'s potential outliers: {outliers}")  
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig, ax = plt.subplots()
ax.set_title("Final Measured Tumor Volume by Drug Regimen")
ax.set_xlabel("Drug Regimen")
ax.set_ylabel("Tumor Vol (mm3)")
ax.boxplot(tumor_vol_list)
plt.xticks([1,2,3,4], treatment_list)

plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
# Create a dataframe with Capomulin Data
campomulin_data = new_index_data.loc[new_index_data["Drug Regimen"] == "Capomulin"]
campomulin_data
```

```python
# Obtain Mean and SE for Capomulin Dataset
stat_campomulin_data = campomulin_data.groupby("Timepoint")["Tumor Volume (mm3)"].agg(['mean','sem'])
stat_campomulin_data
```

```python
# Create a plot
# Create x axis value
x_axis = list(stat_campomulin_data.index.values)

# Make error bar plot with Matplotlib
fig, ax = plt.subplots()
ax.errorbar(x_axis, stat_campomulin_data["mean"], stat_campomulin_data["sem"],
            linestyle='-', marker="o")
ax.set_xlim(-2, 47)
ax.set_ylim(stat_campomulin_data["mean"].min()-2, stat_campomulin_data["mean"].max()+2)
ax.set_xlabel("Time Points (Days)")
ax.set_ylabel("Tumor Volume (mm3)")
ax.set_title("Tumor Volume in 45 Days Trial with Capomulin")

plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
# Generate x and y axis data
id_weight = campomulin_data.drop_duplicates(["Mouse ID", "Weight (g)"], keep='first')
id_weight = id_weight[["Mouse ID", "Weight (g)"]]
avg_campomulin_tumor = campomulin_data.groupby("Mouse ID")["Tumor Volume (mm3)"].mean()

# Create a dataset for scatter plot
merge_for_scatter = pd.merge(id_weight, avg_campomulin_tumor, on=['Mouse ID'], how="left")
merge_for_scatter
```

```python
# Create a scatter plot
plt.scatter(merge_for_scatter["Weight (g)"], merge_for_scatter["Tumor Volume (mm3)"])
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Mouse Weight vs Average Tumor Volume for the Capomulin Regimen")

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# For mouse weight and average tumor volume for the Capomulin regimen

# Correlation
x_values = merge_for_scatter["Weight (g)"]
y_values = merge_for_scatter["Tumor Volume (mm3)"]

correlation = st.pearsonr(x_values, y_values)
print(f"The correlation between mouse weight and average tumor volume is {round(correlation[0],2)}")
```

```python
# Obtain a linear regression model
# Import linregress dependency
from scipy.stats import linregress

x_values = merge_for_scatter["Weight (g)"]
y_values = merge_for_scatter["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(21,38),fontsize=12,color="red")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Mouse Weight vs Average Tumor Volume for the Capomulin Regimen")

print(f"The r-squared is: {round(rvalue**2,2)}")

plt.show()
```
