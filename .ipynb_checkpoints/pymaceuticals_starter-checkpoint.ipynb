## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
# Combine the data into a single dataset
combined_data = pd.merge(study_results, mouse_metadata, on="Mouse ID", how="outer")
```

```python
# Reorganize columns
organized_combined_data = combined_data[["Mouse ID", "Timepoint", "Tumor Volume (mm3)", 
                                         "Metastatic Sites", "Drug Regimen", "Sex", "Age_months", "Weight (g)"]]
organized_combined_data.head()
```

```python
# Sort data with Timepoint
organized_combined_data = combined_data.sort_values(["Timepoint"])
organized_combined_data.head()
```

```python
# Checking the number of mice and dataset
organized_combined_data.describe()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.  
duplicated_data = organized_combined_data[organized_combined_data.duplicated(["Mouse ID", "Timepoint"], keep=False)]

# Optional: Get all the data for the duplicate mouse ID. 
duplicated_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
new_organized_combined_data = organized_combined_data.drop_duplicates(["Mouse ID", "Timepoint"], keep='first')
```

```python
# Checking the number of mice in the clean DataFrame and dataset
new_organized_combined_data.describe()
```

```python
# Reset Index numbers
new_index_data = new_organized_combined_data.reset_index(drop=True)
```

```python
# Display the data table for preview
new_index_data.head()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
means_tumor_volume = new_index_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
median_tumor_volume = new_index_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
var_tumor_volume = new_index_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
std_tumor_volume = new_index_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
SEM_tumor_volume = new_index_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()

# This method is the most straighforward, creating multiple series and putting them all together at the end.
summary_statistics_df = pd.DataFrame({"Mean Tumor Volume": means_tumor_volume, "Median Tumor Volume":median_tumor_volume, 
                                     "Tumor Volume Variance": var_tumor_volume, "Tumor Volume Std. Dev.":std_tumor_volume, 
                                     "Tumor Volume Std. Err.": SEM_tumor_volume})
summary_statistics_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Producing everything in a single groupby function with .agg
alt_summary_statistics_df = new_index_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(['mean', 'median', 'var', 'std', 'sem'])
alt_summary_statistics_df
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pandas. 
total_mice = new_index_data.groupby("Drug Regimen")["Mouse ID"].count()

# Sort total mice data from highest to lowest
sort_total_mice = total_mice.sort_values(ascending = False)

# Make a bar graph
sort_total_mice.plot(kind='bar', width=0.7)

# Put titles for the graph, x and y axis
plt.title("Total Number of Mice for Each Treatment")
plt.xlabel("Drug Regimen")
plt.ylabel("Number of Data Points")

# set x and y axis limit
plt.xlim = (-0.75, 9.75)
plt.ylim = (0, max(total_mice)+10)

plt.show
plt.tight_layout
```

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pyplot.
total_mice = new_index_data.groupby("Drug Regimen")["Mouse ID"].count()
x_axis = total_mice.keys()

# Put titles for the graph, x and y axis
plt.title("Total Number of Mice for Each Treatment")
plt.xlabel("Drug Regimen")
plt.ylabel("Number of Data Points")

# set x and y axis limit
plt.xlim = (-0.75, 9.75)
plt.ylim = (0, max(total_mice)+10)

plt.bar(x_axis, sort_total_mice)
plt.xticks(rotation="vertical")
```
